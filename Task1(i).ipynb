```python
from IPython.display import display
import pandas as pd
```

```python
# Use this if file is already saved in your project folder
df = pd.read_csv("marketing_campaign.csv", sep='\t')  # or sep=',' if comma-separated
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
df = df.dropna(subset=['Income'])
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})
```

```python
df.to_csv("cleaed_marketing_campaign.csv", index=False)
```
